У нас имеется 7043 строки, пропусков нет. Для начала переведем названия столбцов из CamesCase в snake_case.

```python
def camel_to_snake(df):
    column_names = {}
    for column in df.columns.tolist():
        column_names[column] = ''.join('_'+i.lower() if i.isupper() else i for i in column).lstrip('_')
        
    column_names['customerID'] = 'customer_id'
        
    return df.rename(columns=column_names)
```

```python
def get_graph(data, title, xlabel='Наличие услуги', bins=3, ax=None):
    if ax is None:
        data.hist(bins=bins)
        plt.grid=True
        plt.title('Распределение клиентов по ' + title)
        plt.xlabel(xlabel)
        plt.ylabel('Количество клиентов')
        plt.xticks(rotation=0.9)
        plt.show();
    else:
        data.hist(bins=bins, ax=ax)
        ax.grid=True
        
        ax.set_title('Распределение клиентов по ' + title)
        ax.set_xlabel(xlabel)
        ax.set_ylabel('Количество клиентов')
    
```

```python
def get_categorial_graphs(df, categorial_columns):
    cols = 2
    rows = int(np.ceil(len(categorial_columns) / cols))

    plt.figure(figsize=(15, 10))

    for idx, column in enumerate(categorial_columns):
        ax = plt.subplot(rows, cols, idx + 1)

        bins = len(df[column].unique()) * 2 - 1

        get_graph(df[column], column, "", bins, ax=ax)

    plt.show()
```

```python
categorial_columns = ['type', 'paperless_billing', 'payment_method', 
                      'gender', 'senior_citizen', 'partner', 
                      'dependents', 'internet_service', 'multiple_lines']
```

```python
cols = 2
rows = int(np.ceil(len(categorial_columns) / cols))

plt.figure(figsize=(20, 30))

for idx, column in enumerate(categorial_columns):
    grouped_df = df.groupby([column, 'gone'])['gone'].count()
    grouped_df = grouped_df.unstack()
    
    ax = plt.subplot(rows, cols, idx + 1)

    ax.bar(grouped_df.T.index, grouped_df.T[grouped_df.T.columns[0]], label=grouped_df.T.columns[0])
    bottom = np.array(grouped_df.T[grouped_df.T.columns[0]])
    for color_idx, value in enumerate(grouped_df.T.columns[1:]): 
        ax.bar(grouped_df.T.index, grouped_df.T[value], bottom=bottom, label=value)
        bottom += np.array(grouped_df.T[value])
    plt.title(column)
    plt.legend()
        
plt.show()
```

* Люди с помесячным подключением реже уходят. 
* Среди ушедших заметно ниже доля с типом оплаты Mailed Check.
* Распределение по полу одинаково в обеих группах.
* Среди замужних ушедших клиентов чуть больше.
* Среди ушедших доля использующих интернет ниже, но выделяется подключение по оптоволоконному кабелю - таких больше.

```python
features_train_cat, features_test_cat, target_train_cat, target_test_cat = train_test_split(features, target, \
                                                                                  test_size=0.25, random_state=RANDOM_STATE)
```

```python
print(features_train.shape)
print(features_test.shape)
print(target_train.shape)
print(target_test.shape)
```

```python
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
encoder_ohe.fit(features_train[categorial_columns])
features_train[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_train[categorial_columns])
features_train = features_train.drop(categorial_columns, axis=1)

scaler = StandardScaler()
features_train[numeric_columns] = scaler.fit_transform(features_train[numeric_columns])

features_train.head()
```

```python
features_test[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(features_test[categorial_columns])
features_test = features_test.drop(categorial_columns, axis=1)

features_test[numeric_columns] = scaler.transform(features_test[numeric_columns])

features_test.head()
```

## Обучение моделей машинного обучения

### Логистическая регрессия

```python
lr = LogisticRegression(random_state = RANDOM_STATE)

lr_params = {
    'class_weight': (None, 'balanced'),
    'penalty': ('l2', 'l1'),
    'C': (0.1, 0.5, 1, 2, 4)
}

logreg = GridSearchCV(lr, 
                      lr_params, 
                      cv=3, 
                      verbose=2, 
                      scoring='roc_auc',
                      n_jobs = -1)

logreg.fit(features_train, target_train)
```

```python
logreg.best_score_
```

```python
logreg.best_params_
```

### Random Forest

```python
rf = RandomForestClassifier(random_state=RANDOM_STATE)
forest_params = {
    'max_depth': (5, 7, 10, 15),
    'n_estimators': (100, 150, 300),
    'class_weight': (None, 'balanced')
}

forest = GridSearchCV(rf, 
                      forest_params, 
                      cv=3, 
                      verbose=2, 
                      scoring='roc_auc',
                      n_jobs = -1)

forest.fit(features_train, target_train)
```

```python
forest.best_score_
```

```python
forest.best_params_
```

### CatBoost

```python
catboost = CatBoostClassifier(random_state=RANDOM_STATE, verbose=False)


cat_params = {
      'learning_rate': [0.1, 0.2, 0.3, 0.5],
      'depth': [3, 5, 7],
      'num_trees': (150, 300, 450),
}

cat = GridSearchCV(catboost, 
                      cat_params, 
                      cv=3, 
                      verbose=2, 
                      scoring='roc_auc',
                      n_jobs = -1)

cat.fit(features_train_cat, target_train_cat, cat_features=categorial_columns)
```

```python
cat.best_score_
```

```python
cat.best_params_
```

## Выбор лучшей модели

### Анализ моделей

```python
df = pd.DataFrame([['Logistic Regression', round(logreg.best_score_, 3)], 
                   ['RF', round(forest.best_score_, 3)],
                   ['CatBoost', round(cat.best_score_, 3)]],
                   columns=['Model', 'ROC-AUC on CV'])
df.sort_values(by='ROC-AUC on CV', ascending=False)
```

Наилучший результат показала модель CatBoost с параметрами {'depth': 3, 'learning_rate': 0.3, 'num_trees': 450}. Ее и будем проверять на тестовой выборке.

### Проверка лучшей модели на тестовой выборке

```python
predictions = cat.best_estimator_.predict(features_test_cat)
probabilities_test = cat.best_estimator_.predict_proba(features_test_cat)
probabilities_one_test = probabilities_test[:, 1]

print('ROC-AUC:', roc_auc_score(target_test_cat, probabilities_one_test))
print('Accuracy:', accuracy_score(target_test_cat, predictions))
```

```python
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)
plt.figure()
plt.plot(fpr, tpr, linestyle='--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()
```

Модель делает мало ошибок первого, но хуже справляется с ошибками второго рода - 6% клиентов, которые уйдут, мы не сможем распознать.
