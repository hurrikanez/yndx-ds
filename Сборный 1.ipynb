**Описание проекта**

Вы работаете в интернет-магазине, специализирующемся на продаже компьютерных игр по всему миру. У вас есть доступ к историческим данным о продажах игр, пользовательских и экспертных оценках, жанрах и платформах (например, Xbox или PlayStation) из открытых источников. Ваша задача заключается в выявлении факторов, определяющих успешность игры. Это поможет сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Предоставленные данные охватывают период до 2016 года, и предположим, что наступил декабрь 2016 года, и вы планируете кампанию на 2017 год. Важно научиться работать с данными, независимо от того, прогнозируете ли вы продажи на 2017 год на основе данных 2016 года или же на 2027 год — на основе данных 2026 года.

В предоставленном наборе данных встречается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, такие как "Для взрослых", "Для детей младшего возраста" или "Для подростков".

**Описание данных**

* *Name* — название игры
* *Platform* — платформа
* *Year_of_Release* — год выпуска
* *Genre* — жанр игры
* *NA_sales* — продажи в Северной Америке (миллионы проданных копий)
* *EU_sales* — продажи в Европе (миллионы проданных копий)
* *JP_sales* — продажи в Японии (миллионы проданных копий)
* *Other_sales* — продажи в других странах (миллионы проданных копий)
* *Critic_Score* — оценка критиков (максимум 100)
* *User_Score* — оценка пользователей (максимум 10)
* *Rating* — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

Со столбцом *user_score* ситуация наиболее сложная. Пропусков там больше всего, еще и тип данных - строковый, хотя значения в нем определенно числовые. Строковым его делает значение tbd.
> TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято.

Предположим, что это значение проставлено у недавно вышедших игр, еще не успевших получить оценку пользователя.

Предположение оказалось неверно - такая оценка есть у игр разных годов выпуска. <br>
Также можно предположить, что это значение проставляется у игр с недосаточным для высчитывания адекватной оценки количеством оценок пользователя. Эту теорию мы проверить, увы, не сможем.<br>
Для того, чтобы далее работать с числовыми значениями в этом столбце, заменим значение tbd на NaN. В рамках требуемых результатов для нас одинаково отсутствие значения или его неопределенность.

#### Влияние оценок на продажи
Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. 
Для анализа возьмем PS4.

```python
ps4_games = latest_games.query('platform == "PS4"')
```

```python
ps4_games.plot(kind='scatter', x='critic_score', y='total_sales', title='Связь оценок критиков и продаж для PS4')
plt.xlabel('Оценка критиков')
plt.ylabel('Продажи, млн. копий')
ps4_games.plot(kind='scatter', x='user_score', y='total_sales', title='Связь оценок пользователей и продаж для PS4');
plt.xlabel('Оценка пользователей')
plt.ylabel('Продажи, млн. копий');
```

На примере PS4 можно заметить, что связь между оценкой критиков и продажами более сильная, чем между оценкой пользователей и продажами. При этом высокая оценка что у тех, что у других, не является гарантией высоких показателей продаж. <br>
Значения на графике с оценкой пользователя намного более рассеянные - это может быть связано с тем, что геймеры более привередливая аудитория, и могут быть более субъективны, чем критики. Кроме того, критики могут выпускать обзоры на игры еще до их официального выхода или в самом начале, влияя на продажи, а отзывы игроков появляются уже позже. <br>
Основываясь на коэффициенте кореляции Пирсона можно сказать, что в случае оценки критиков связь является нелинейной, и продажи зависят не только от оценок, но и от других показателях. Для оценки пользователей коээфициент показывает, что связь вообще скорее отсутствует.

### Портрет пользователей

Определим для пользователя каждого региона (NA, EU, JP):
* Самые популярные платформы (топ-5). Опишем различия в долях продаж.

* Влияет ли рейтинг ESRB на продажи в отдельном регионе?

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']

def top_in_region(category, region, ax):
    df = latest_games.pivot_table(index=category, values=region, aggfunc='sum').\
        sort_values(by=region, ascending=False).reset_index()
    if len(df) > 5:
        others = {category: 'Others', region: df[5:][region].sum()}
        df = df[:5].append(others, ignore_index=True)
    df = df.set_index(category)
    
    df.plot(kind='pie', y=region, ylabel='', autopct='%1.0f%%', title=region[:2].upper(), ax=ax);
    ax.legend(bbox_to_anchor=(0.9, 1), loc='upper left', borderaxespad=0);
```

### Проверка гипотез

Нужно проверить следующие гипотезы:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Данные в обоих случах независимы, а выборочные средние распределены нормально вокруг истинного среднего, так как размеры выборки составляют более нескольких десятков, поэтому мы можем использовать t-тест.

Также нам нужно выбрать p-value - вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза будет верна. Общепринятые пороговые значения — 5% и 1%. Возьмем 5% - то есть, если наша нулевая гипотеза верна, вероятность увидеть значения, которые мы получим, будет не менее 5%.


#### Xbox One и PC
Начнем с первой.<br>
Нулевая гипотеза (H0) будет звучать так: Средние пользовательские рейтинги платформ Xbox One и PC равны.<br>
Альтернативная (H1) будет такой: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
pc_games = latest_games.query('platform == "PC"')
# xone_games уже была создана ранее

print(len(pc_games))
print(len(xone_games))
```

Размеры выборок значительно отличаются, дисперсии выборок будем считать как разные.

```python
pc_user_score = pc_games[pc_games['user_score'].notna()]['user_score']
xone_user_score = xone_games[xone_games['user_score'].notna()]['user_score']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(pc_user_score, xone_user_score, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

По полученным результатам нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC не отвергаем.

#### Action и Sports
Займемся следующей гипотезой. <br>
Н0 (Нулевая): Средние пользовательские рейтинги жанров Action и Sports равны. <br>
H1 (Альтернативная): Средние пользовательские рейтинги жанров Action и Sports не равны.

```python
action_games = latest_games.query('genre == "Action"')
sports_games = latest_games.query('genre == "Sports"')

print(len(action_games))
print(len(sports_games))
```

Размеры выборок значительно отличаются, дисперсии выборок будем считать как разные.

```python
action_games_score = action_games[action_games['user_score'].notna()]['user_score']
sports_games_score = sports_games[sports_games['user_score'].notna()]['user_score']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(action_games_score, sports_games_score, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
action_games_score.mean()
```

```python
sports_games_score.mean()
```

По полученным результатам следует отвергнуть гипотезу о равенстве средних пользовательских рейтингов игр жанров Action и Sports - значение p-value *очень* мало. Посчитав средние, можно увидеть, что средний рейтинг у жанра Action выше.

**Вывод**: в результате проверок статистических гипотез можно говорить о том, что средние пользовательские рейтинги платформ Xbox One и PC с достаточной вероятностью одинаковые, а средние пользовательские рейтинги жанров Action и Sports нельзя считать равными.

* Таким образом, при планировании кампании следует учитывать, что рынок Японии в целом значительно отличается от рынка Европы и Америки, и кампании для них следуют планировать разные. 
* Можно говорить о том, что средние пользовательские рейтинги платформ Xbox One и PC с достаточной вероятностью одинаковые, а средние пользовательские рейтинги жанров Action и Sports нельзя считать равными - у Action он выше.
