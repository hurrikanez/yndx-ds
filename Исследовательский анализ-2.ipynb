# Исследование объявлений о продаже квартир

У вас есть доступ к данным сервиса Недвижимости, который содержит информацию о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Ваша цель — определить рыночную стоимость недвижимости и установить соответствующие параметры. Это позволит создать автоматизированную систему, способную выявлять аномалии и мошенническую деятельность. 

Для каждой квартиры доступны два вида данных: первые вводятся пользователем, вторые получены автоматически на основе картографических данных, таких как расстояние до центра, аэропорта, ближайшего парка и водоёма.

**Описание данных**

* *airports_nearest* — расстояние до ближайшего аэропорта в метрах (м)
* *balcony* — число балконов
* *ceiling_height* — высота потолков (м)
* *cityCenters_nearest* — расстояние до центра города (м)
* *days_exposition* — сколько дней было размещено объявление (от публикации до снятия)
* *first_day_exposition* — дата публикации
* *floor* — этаж
* *floors_total* — всего этажей в доме
* *is_apartment* — апартаменты (булев тип)
* *kitchen_area* — площадь кухни в квадратных метрах (м²)
* *last_price* — цена на момент снятия с публикации
* *living_area* — жилая площадь в квадратных метрах (м²)
* *locality_name* — название населённого пункта
* *open_plan* — свободная планировка (булев тип)
* *parks_around3000* — число парков в радиусе 3 км
* *parks_nearest* — расстояние до ближайшего парка (м)
* *ponds_around3000* — число водоёмов в радиусе 3 км
* *ponds_nearest* — расстояние до ближайшего водоёма (м)
* *rooms* — число комнат
* *studio* — квартира-студия (булев тип)
* *total_area* — общая площадь квартиры в квадратных метрах (м²)
* *total_images* — число фотографий квартиры в объявлении

* Затем уже идут деревни, пгт и сёла - застройка там снова малоэтажная. Туда отнесем всё, что больше 20 км.

Создадим функцию для группировки и заполним пропуски в этих столбцах медианным значением по категориям.

```python
def categorize_house(row):
    if row['city_centers_nearest'] < 4 * 1000:
        return 'old'
    elif row['city_centers_nearest'] < 10 * 1000:
        return 'middle'
    elif row['city_centers_nearest'] < 20 * 1000:
        return 'new'
    elif row['city_centers_nearest'] >= 20 * 1000:
        return 'suburbian'
    else:
        return 'unknown'
```

```python
real_estate['house_category'] = real_estate.apply(categorize_house, axis=1)
```

```python
real_estate.sample(10) # проверим, что отработало корректно
```

Посмотрим, есть ли действительно разница по категориям.

```python
real_estate.groupby('house_category')['ceiling_height'].median().round(2)
```

```python
real_estate.groupby('house_category')['floors_total'].median().round()
```

Наглядно видно, что самые высокие потолки - в старых домах и домах застройки второй половины 20 века (видимо, много сталинок). В остальных категориях высота самая стандартная - около 2,6 метра.
Как и предпологалось, этажность меньше в центре, возрастает в районах новостроек, а в других частях города более популярны 5-этажки.

В категории 'unknown' (там, где мы не знаем расстояние до центра) значения получились достаточно средние. Хорошо подойдет для заполнения в таком варианте.

Теперь нужно заполнить пропуски.

```python
for category in real_estate['house_category'].unique():
    filtered_houses = real_estate.loc[real_estate['house_category'] == category] # отфильтрованные по категории
        
    # вставляем в пропуски среднюю высоту потолков
    real_estate.loc[real_estate['house_category'] == category, 'ceiling_height'] = \
        real_estate.loc[real_estate['house_category'] == category, 'ceiling_height'] \
        .fillna(filtered_houses['ceiling_height'].median().round(2))
    # вставляем в пропуски количество этажей в доме
    real_estate.loc[real_estate['house_category'] == category, 'floors_total'] = \
        real_estate.loc[real_estate['house_category'] == category, 'floors_total'] \
        .fillna(filtered_houses['floors_total'].median().round(0))   
    
# если этаж квартиры вдруг оказался больше среднего, правим     
real_estate['floors_total'] = real_estate[['floor', 'floors_total']].max(axis=1)
```

Разберемся с пропусками в *living_area* и *kitchen_area*.

Заполнить пропуски в *living_area* медианным или средним значением по столбцу будет не совсем корректно, так как для маленьких квартир это значение может слишком приблизиться к общей площади квартиры.
К жилой площади относятся только те помещения, которые пригодны для проживания: гостиная, спальня, детская. Сюда не относятся: кухня, санузлы, коридор, кладовые и гардеробные. Значит, у нас еще есть место для погрешности в виде санузлов и корридоров.

Посчитаем медианное соотношение общей площади квартиры и жилой. У однокомнатной и четырехкомнатной квартир эти соотношения будут разными. Поэтому и считать, и заполнять будем с соответствии с количеством комнат в квартире. В итоге, пропуски заполняем по формуле *медианное соотношение жилой площади к общей у квартир с таким количеством комнат* умножить на *общую площадь квартиры*. 

То же самое сделаем и для кухонь, но с одной поправкой. Жилая площадь студии равна суммарной (общей) площади минус квадратные метры санузла (это мы нагуглили). Значит, кухонная зона тоже считается жилой. Для студий оставим в *kitchen_area* пропуски.

Наиболее вероятной причиной пропусков в этих столбцах здесь может быть человеческий фактор (*владельцу лень считать/вспоминать сколько там, общей площади достаточно*).

```python
area_dict = {'rooms': [], 'living_ratio': [], 'kitchen_ratio': []}

for room_number in real_estate['rooms'].unique():
    # для жилой площади
    filtered_rooms = real_estate.loc[real_estate['rooms'] == room_number] # отфильтрованные по количеству комнат квартиры
    living_ratio = (filtered_rooms['living_area'] / filtered_rooms['total_area']).median() # считаем соотношение для жилой
    
    # вставляем в пропуски жилой площади соотнешние умноженное на общую площадь квартиры
    real_estate.loc[real_estate['rooms'] == room_number, 'living_area'] = \
        real_estate.loc[real_estate['rooms'] == room_number, 'living_area'] \
        .fillna(filtered_rooms['total_area'] * living_ratio)
    # для кухонь
    if room_number != 0:
        kitchen_ratio = (filtered_rooms['kitchen_area'] / filtered_rooms['total_area']).median() # считаем соотношение для кухни
    else:
        kitchen_ratio = 0
    
    # вставляем в пропуски площади кухни соотнешние умноженное на общую площадь квартиры
    real_estate.loc[real_estate['rooms'] == room_number, 'kitchen_area'] = \
        real_estate.loc[real_estate['rooms'] == room_number, 'kitchen_area'] \
        .fillna(filtered_rooms['total_area'] * kitchen_ratio)

        
    area_dict['rooms'].append(room_number)
    area_dict['living_ratio'].append(living_ratio)
    area_dict['kitchen_ratio'].append(kitchen_ratio)
```

### Посчитайте и добавьте в таблицу новые столбцы

Итак, нам нужно добавить в таблицу новые столбцы со следующей информацией:
* *m2_price* - цена одного квадратного метра;
* *weekday_exposition* - день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
* *month_exposition* - месяц публикации объявления;
* *year_exposition* - год публикации объявления;
* *floor_type* - тип этажа квартиры (значения — «первый», «последний», «другой»);
* *city_centers_km* - расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).

```python
# создаем столбец цены 1 квадратного метра - сразу округлим до целых и приведем к целочисленному типу.
real_estate['m2_price'] = (real_estate['last_price'] / real_estate['total_area']).round().astype('int')
```

```python
# создаем столбец дня недели публикации
real_estate['weekday_exposition'] = real_estate['first_day_exposition'].dt.day_of_week
```

```python
# создаем столбец с месяцем публикации
real_estate['month_exposition'] = real_estate['first_day_exposition'].dt.month
```

```python
# создаем столбец с годом публикации
real_estate['year_exposition'] = real_estate['first_day_exposition'].dt.year
```

```python
# создаем функцию для типирования этажа
def type_floor(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
# типируем этаж
real_estate['floor_type'] = real_estate.apply(type_floor, axis=1)
```

```python
# создаем столбец с расстоянием до центра в километрах, оставим его в типе float, так как в столбце с расстоянием остались NaN
real_estate['city_centers_km'] = (real_estate['city_centers_nearest'] / 1000).round()
```
